Our first approach is to use bagging with decision trees.

```python
# let's make the basic imports

import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['figure.dpi']= 180 # set the resolution to x dpi

data = pd.read_csv("test_cleaned.csv")
# data = pd.read_csv("finalDataSet.csv")
data.describe()
```

```python
# We split the data into training and testing
from sklearn.model_selection import train_test_split

test_size = 0.2
train, test = train_test_split(data,test_size=test_size,shuffle=True)
```

```python
print(train.shape,test.shape)
print(np.sum(np.sum(train.isnull())),np.sum(np.sum(train.isnull())))
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    "criterion":["gini","entropy"],
    "max_depth" : list(range(2,train.shape[1]-1,2)) + [None]}

classifier = GridSearchCV(DecisionTreeClassifier,param_grid,n_jobs=-1)
features,target = (train.iloc[:,:-1],train.iloc[:,-1].astype(int))
print(features.shape,target.shape)
classifier.fit(features,target.ravel())
```
